# Описание проекта

В нашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Наша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

**Описание столбцов:**
- `airports_nearest` — расстояние до ближайшего аэропорта в метрах (м)
- `balcony` — число балконов
- `ceiling_height` — высота потолков (м)
- `cityCenters_nearest` — расстояние до центра города (м)
- `days_exposition` — сколько дней было размещено объявление (от публикации до снятия)
- `first_day_exposition` — дата публикации
- `floor` — этаж
- `floors_total` — всего этажей в доме
- `is_apartment` — апартаменты (булев тип)
- `kitchen_area` — площадь кухни в квадратных метрах (м²)
- `last_price` — цена на момент снятия с публикации
- `living_area` — жилая площадь в квадратных метрах (м²)
- `locality_name` — название населённого пункта
- `open_plan` — свободная планировка (булев тип)
- `parks_around3000` — число парков в радиусе 3 км
- `parks_nearest` — расстояние до ближайшего парка (м)
- `ponds_around3000` — число водоёмов в радиусе 3 км
- `ponds_nearest` — расстояние до ближайшего водоёма (м)
- `rooms` — число комнат
- `studio` — квартира-студия (булев тип)
- `total_area` — площадь квартиры в квадратных метрах (м²)
- `total_images` — число фотографий квартиры в объявлении

Значений очень много и не помешало бы как их сократить, возможно населенные пункты названы по разному.

Попробуем как-нибудь разделить тип и название населенного пункта. Все слова начинающиеся с большое буквы в 1 столбец, а остальное в другой, для этого сделаем функции с помощью которой будем обрабатывать каждую строку.

```python
def name_loc_ABC(row):
    list_elem = str(row['locality_name']).split()
    list_resul = []
    for elem in list_elem:
        if elem[0].isupper():
            list_resul.append(elem)
    return ' '.join(list_resul)
```

```python
def name_loc_abc(row):
    list_elem = str(row['locality_name']).split()
    list_resul = []
    for elem in list_elem:
        if 'ё' in elem:
            elem = elem.replace('ё','е')
        if elem[0].islower():
            list_resul.append(elem)
    if len(list_resul) == 0:
        return 'город'
    return ' '.join(list_resul)
```

```python
df['locality_type'] = df.apply(name_loc_abc, axis=1)
```

```python
df['locality_name'] = df.apply(name_loc_ABC, axis=1)
```

Как видно заполненные значение есть у 145 населенных пункта. Для них тогда и заполним пустые значения, остальные оставим без изменения, так как возможно это указывает на большую удаленность от аэропорта. Так же учетем, что в некоторых населенных пунктах вообще нет значения расстояния до аэропорта.
Для заполнения применим цикл:

```python
list_locality = df[df['airports_nearest'].isna()]['locality_name'].unique()

for name in list_locality:
    df_query_name = df.query('locality_name == @name')['airports_nearest']
    # Найдем теперь где есть хотя бы 1 значение расстояния до аэропорта
    # и заполним медианой пропуски
    if df_query_name.shape[0] != df_query_name.isna().sum():
        median_query_name = df.query('locality_name == @name')['airports_nearest'].median()
        df.loc[ (df['locality_name']==name) & 
                (df['airports_nearest'].isna()),
                'airports_nearest'
                ] = median_query_name
        print('Обработан нас. пункт:', name)
        print(f'Обработано {df_query_name.isna().sum()} значений')
```

Так же же нужно не забывать тот факт что в данных есть значения общей площади менее 22, а значение жилой площади не может быть 0 или отрицательным числом.

В данных имеются значения с странно маленькой жилой площадью, об этой особенности я бы посоветовался с коллегами на предмет адекватности значений. 

Попробуем сделать плавный переход, чем меньше общая площадь помещения, тем меньше будет разница между общей и жилой площадью. Для этого напишем функцию.

```python
def fill_na_liv_area (row):
    total_area = row['total_area']
    liv_area = row['living_area']
    if math.isnan(liv_area):
        if total_area <= 22:
            return total_area - 2
        elif total_area <= 32:
            return total_area - 12
        else:
            return total_area - 22
    return liv_area
```

### Цены на основе расстояния до центра

Посмотрим более глубже на зависимость цены от расстояния до центра. Сделаем новую колонку, где расстояние будет вычесленно в километрах.

```python
df['cityCenters_nearest_km'] = round(df['cityCenters_nearest'] / 1000)
df['cityCenters_nearest_km'] = df['cityCenters_nearest_km'].astype('Int16')
df['cityCenters_nearest_km'].head()
```

Посмотрим среднюю цену на каждом километре от центра и построим график

```python
flat_price = df.groupby('cityCenters_nearest_km')['last_price'].mean()
flat_price.plot()

plt.title('Графки цены от расстояния до центра')
plt.xlabel('Удаленность от центра, км')
plt.ylabel('Цена квартиры, тыс. руб')
plt.show()
```

Как можно увидеть на графике, в районе 5 километров находится "полочка" значений. Скорей всего там и заканчивается определение центра

```python
flat_price[3:10]
```

Будем считать что центр заканчивается на 8 километре

```python
df_center = df[df['cityCenters_nearest_km'] < 8]
df_center.shape
```

Объявлений с квартирой в центре Питера оказалось 2567 штук.  

Посмотрим какие закономерности есть в данных квартирах и отличаются ли они от закономерностей в общей таблице

```python
sns.pairplot(data=df_center[['last_price', 'total_area', 'rooms', 'cityCenters_nearest']],
             x_vars=['last_price', 'total_area', 'rooms', 'cityCenters_nearest'],
             y_vars=['last_price']
            )
plt.show()
```

```python
df_center[['last_price', 'total_area', 'rooms', 'cityCenters_nearest']].corr()
```

Исходя из графиков можем заметить что цена в центре распределена нормально, видится линейная корреляция цены и общей площади как и в общей таблице, единственное что с уменьшением общей площади видно четкая коррелция площади со стоимостью, разброс цены так уменьшается (конусообразный график). При увеличении цены сокращаются варианты с количеством комнат и как правило у более дорогих квартир нет вариантов с маленьким количеством комнат и чем больше цена, тем боольшей минимальнок количество комнат в выборе. Все объявления в центре размещены хаотично, тоесть на цену никак не влияет удаленность от точки взятого как центр города. Так же можно заметить в непосредственной близости от центра имеются малое количество дешевых квартит (дешевле 5 млн руб.), но это применино условно до 2 километров.

```python
sns.boxplot(data=df_center, x='year_exposition' , y= 'last_price')
plt.title('Сравнение цены объявлений по годам')
plt.xlabel('Год размещения объявления')
plt.ylabel('Цена крартиры, тыс. руб')
plt.show()
```
